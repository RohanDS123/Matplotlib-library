# Bar plot word cup 2023.

```python
permision = input("Enter the your permision here:")
if permision=="you can run this program":
    x = ["Virat kohli","Quinton de Kock","Rachin Ravindra","Rohit Sharma"]
    y = [594,591,565,503]
    plt.xlabel("Players name",fontsize = 15,color = "y")
    plt.ylabel("Run level",fontsize = 15,color = "y")
    plt.title("WCP TRR till 14 november",fontsize = 18,color = "y")
    c = ["r","g","y","k"]
    plt.bar(x,y,color = c,width = 0.4,align = "center",edgecolor = "k",linewidth = 3,label = "World cup 2023")
    plt.legend()
    plt.xticks(range(len(x)), x, fontsize=10, rotation=45, ha="right",color="darkred")
    plt.ylim(400,600)   # ylim means set the limit of y axis.
    plt.show()
else:
    print("Your program will not be run because it is  not given to permision")
```

# Multiple bar plot -

```python
permision = input("Enter the your permision here:")
if permision=="you can run this program":
    width = 0.3
    x = ["Virat kohli","Quinton de Kock","Rachin Ravindra","Rohit Sharma"]
    y = [594,591,565,503]
    z = [443,447,566,456]
    p = np.arange(len(x))
    p1 = [j + width for j in p]
    plt.xlabel("Players name",fontsize = 15,color = "y")
    plt.ylabel("Run level",fontsize = 15,color = "y")
    plt.title("WCP TRR till 14 november",fontsize = 18,color = "y")
    width = 0.3
    plt.bar(p,y,color = "r",width = 0.3,label = "World cup 2023")
    plt.bar(p1,z,color = "k",width = 0.3,label = "World cup 2023")
    plt.legend()
    plt.xticks(range(len(x)), x, fontsize=10, rotation=45, ha="right",color="darkred")
    plt.ylim(400,600)   # ylim means set the limit of y axis.
    plt.show()
else:
    print("Your program will not be run because it is  not given to permision")
```

# Scatter plot  - Sachin Tendulkar Run list

```python
x = [200,463,1,310]
y = [51921,18426,10,25396]
plt.xlabel("All match",fontsize = 16,color = "y")
plt.ylabel("Runs",fontsize = 16,color = "y")
plt.title("All match  and runs Sachin Tendulkar",fontsize = 17,color = "Purple")
plt.scatter(x,y,marker = "*",cmap = "PuRd",s = 440,edgecolor = "r",linewidth = 3)
plt.legend()
t = plt.colorbar()
t.set_label("colorbar",color = "Purple",fontsize = 13)
plt.show()
```

# Pie chart --  Cristiano Ronaldo

```python
x = [450,5,127]
y = ["Real Madrid","Sporting CP","Portugal"]
plt.title("Pie chart")
c = ["Purple","r","b"]
ex = [0.0,0.0,0.0]
plt.pie(x,labels=y,colors = c,explode = ex,shadow = True,radius = 1,autopct = "%0.1f%%",labeldistance = 1.1
,textprops = {"fontsize":12,"color":"r"},wedgeprops = {"linewidth": 4, "edgecolor": "darkred"}
)
plt.show()
```

 # Histogram plot - 

```python
x = np.random.randint(10,60,(50))
```

```python
x
```

```python
n = [43, 30, 25, 31, 50, 32, 43, 30, 40, 48, 39, 26, 54, 38, 11, 35, 44,
       59, 31, 22, 39, 47, 58, 59, 11, 40, 37, 24, 48, 57, 23, 18, 43, 45,
       31, 48, 28, 32, 46, 16, 43, 54, 19, 30, 27, 34, 50, 30, 45, 46]
```

```python
plt.xlabel("Exact value",fontsize = 15,color = "y")
plt.ylabel("Given range",fontsize = 15,color = "y")
plt.title("Histogram plot",fontsize = 16,color = "r")
plt.hist(n,cumulative = -1,color = "k",edgecolor = "r",linewidth = 3,align = "left")
plt.show()
```

# All the types of histograms - 

__This is a step histogram__

```python
plt.xlabel("Exact value",fontsize = 15,color = "y")
plt.ylabel("Given range",fontsize = 15,color = "y")
plt.title("Histogram plot",fontsize = 16,color = "r")
plt.hist(n,cumulative = -1,color = "k",edgecolor = "r",linewidth = 3,align = "left",histtype = "step")
plt.show()
```

__This is a stepfilled histogram__

```python
plt.xlabel("Exact value",fontsize = 15,color = "y")
plt.ylabel("Given range",fontsize = 15,color = "y")
plt.title("Histogram plot",fontsize = 16,color = "r")
plt.hist(n,cumulative = -1,color = "Purple",edgecolor = "r",linewidth = 3,align = "left",histtype = "stepfilled")
plt.show()
```

__This is a barstacked histogram__

```python
plt.xlabel("Exact value",fontsize = 15,color = "y")
plt.ylabel("Given range",fontsize = 15,color = "y")
plt.title("Histogram plot",fontsize = 16,color = "r")
plt.hist(n,cumulative = -1,color = "k",edgecolor = "r",linewidth = 3,align = "left",histtype = "barstacked")
plt.show()
```

__Convert this histogram verticle to horizontal__

```python
plt.xlabel("Exact value",fontsize = 15,color = "y")
plt.ylabel("Given range",fontsize = 15,color = "y")
plt.title("Histogram plot",fontsize = 16,color = "r")
plt.hist(n,cumulative = -1,color = "k",edgecolor = "r",linewidth = 3,align = "left",orientation = "horizontal")
plt.show()
```

```python
plt.xlabel("Exact value",fontsize = 15,color = "y")
plt.ylabel("Given range",fontsize = 15,color = "y")
plt.title("Histogram plot",fontsize = 16,color = "r")
plt.hist(n,cumulative = -1,color = "k",edgecolor = "r",linewidth = 3,align = "left",rwidth = 0.8)
plt.show()
```

__Converting to this table into the log table__

```python
plt.xlabel("Exact value",fontsize = 15,color = "y")
plt.ylabel("Given range",fontsize = 15,color = "y")
plt.title("Histogram plot",fontsize = 16,color = "r")
plt.hist(n,cumulative = -1,color = "k",edgecolor = "r",linewidth = 3,align = "left",log = True)
plt.show()
```

__Converting to this into the like a boxes__

```python
plt.xlabel("Exact value",fontsize = 15,color = "y")
plt.ylabel("Given range",fontsize = 15,color = "y")
plt.title("Histogram plot",fontsize = 16,color = "r")
plt.hist(n,cumulative = -1,color = "k",edgecolor = "r",linewidth = 3,align = "left")
plt.grid()
plt.show()
```
